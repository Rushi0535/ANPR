```python
video = cv2.VideoCapture('videos/video-9.mp4')

object_detector = cv2.createBackgroundSubtractorMOG2(history=100, varThreshold=75)

detect = []

while True:
    ret, frame = video.read()
    
    grey = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(grey, (5,5), 3)
    #object detection
    mask = object_detector.apply(blur)
    _, mask = cv2.threshold(mask, 254, 255, cv2.THRESH_BINARY)
    
    dilate = cv2.dilate(mask, None, iterations=3)
    kernel = np.ones((5,5), np.uint8)
    dilate= cv2.morphologyEx(dilate, cv2.MORPH_CLOSE, kernel)
    dilate = cv2.morphologyEx(dilate, cv2.MORPH_CLOSE, kernel)
    contours, _ = cv2.findContours(dilate, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    dilate_resize = cv2.resize(dilate, (600,400))
   
    cv2.line(frame, (25, 550), (1200, 550), (255,0,0), 5)
    
    for (i,c) in enumerate(contours):
        (x, y, w, h) = cv2.boundingRect(c)
        validate_contour = (w >= 80) and (h >= 80)
        if not validate_contour:
            continue
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 3)
        x1 = int(w/2)
        y1 = int(h/2)
        cx = x + x1
        cy = y + y1
        cv2.circle(frame, (cx, cy), 5, (0, 0, 255), -1)
        detect.append([cx, cy])

    frame = cv2.resize(frame, (600,400))
    cv2.imshow('mask',dilate_resize)
    cv2.imshow('frame', frame)
    

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

video.release()
cv2.destroyAllWindows()
```
